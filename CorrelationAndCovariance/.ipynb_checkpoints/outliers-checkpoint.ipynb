```python
# In statistics, a z-score tells us how many standard deviations away a value is from the mean. We use the following formula to calculate a z-score:
# z = (X – μ) / σ
# where
# X is a single raw data value
# μ is the population mean
# σ is the population standard deviation
```

```python
def plot_hist(p_col_name,p_col_data,p_min_limit_to_outlier,p_max_limit_to_outlier,p_total):
    #col_data.plot.hist(title=p_col_name+'(Range '+str(p_min_limit_to_outlier)+' to '+str(p_max_limit_to_outlier)+')',alpha=0.5,bins=100)
    #plt.show()
    print(p_col_name,'\t\tTotal of outliers: ',p_total,' Valid interval [%.3f to %.3f]' %(p_min_limit_to_outlier,p_max_limit_to_outlier))
```

```python
"""
This first method performs the filter zipping the z and the respective value
"""
def met_1(p_col_name,p_col_data,p_threshold):
    z=stats.zscore(p_col_data)
    min_limit_to_outlier=np.mean(p_col_data)
    max_limit_to_outlier=np.mean(p_col_data)
    total_valid=0
    for vz,vd in zip(z,p_col_data):
        if vz>0 and vz<p_threshold:
            total_valid=total_valid+1
            if vd>max_limit_to_outlier:
                max_limit_to_outlier=vd
        if vz<0 and abs(vz)<p_threshold:
            total_valid=total_valid+1
            if vd>min_limit_to_outlier:
                min_limit_to_outlier=vd
    total=len(p_col_data)-total_valid
    plot_hist(p_col_name,p_col_data,min_limit_to_outlier,max_limit_to_outlier,total)

```

```python
"""
The second method uses the np.where in order to obtain the indexes of the outliers
"""
def met_2(p_col_name,p_col_data,p_threshold):
    z=stats.zscore(p_col_data)
    # The z_outliers is a tuple with an array containing the indexes
    z_outliers_lower=np.where(z < p_threshold*-1)
    z_outliers_upper=np.where(z > p_threshold)
    total_lower=len(z_outliers_lower[0])
    total_upper=len(z_outliers_upper[0])
    total=total_lower+total_upper
    #TODO: considerar os válidos e não os outliers
    #valid_data=np.delete(p_col_data, z_outliers_upper[0], axis=None)
    min_limit_to_outlier=np.max(p_col_data[z_outliers_lower[0]]) if total_lower>0 else np.min(col_data)
    max_limit_to_outlier=np.min(p_col_data[z_outliers_upper[0]]) if total_upper>0 else np.max(col_data)
    plot_hist(p_col_name,p_col_data,min_limit_to_outlier,max_limit_to_outlier,total)
```
