```python
"""
This method performs the filter by zipping the z (z-score) and the respective value
"""
def analysis_zscore(p_col_name,p_col_data,p_threshold,p_verbose):
    z=stats.zscore(p_col_data)
    min_limit_to_outlier=np.max(p_col_data)
    max_limit_to_outlier=np.mean(p_col_data)
    total_valid=0
    total_outliers_lower=0
    total_outliers_upper=0
    # Search the valid values with a z-score above the limit
    for vz,vd in zip(z,p_col_data):
        if vz>=0:
            if vz<p_threshold:
                # The z-score is positive, but the value is still valid
                total_valid=total_valid+1
                if vd>max_limit_to_outlier:
                    max_limit_to_outlier=vd
            else:
                total_outliers_upper=total_outliers_upper+1
                if p_verbose:
                    print('  ',p_col_name+'  Upper outlier: ',vd,'  z=',vz)
        else:
            if abs(vz)<p_threshold:
                # The z-score is negative, but the value is still valid
                total_valid=total_valid+1
                if vd<min_limit_to_outlier:
                    min_limit_to_outlier=vd
            else:
                total_outliers_lower=total_outliers_lower+1
                if p_verbose:
                    print('  ',p_col_name+'  Lower outlier: ',vd,'  z=',vz)
    
    if total_outliers_lower==0:
        min_limit_to_outlier=np.min(p_col_data)
    if min_limit_to_outlier==0:
        max_limit_to_outlier=np.max(p_col_data)
    if p_verbose:
        print(p_col_name,'\t\tTotal of outliers: ',total_outliers_lower,'+',total_outliers_upper,'  Valid interval [%.3f to %.3f]' %(min_limit_to_outlier,max_limit_to_outlier))
    title='%s\t- Z Range: %.3f to %.3f \t Total of Outliers: %d and %d'%(p_col_name,min_limit_to_outlier,max_limit_to_outlier,total_outliers_lower,total_outliers_upper)
    return title
```

```python
from sklearn.linear_model import LinearRegression
def analysis_LinearRegression(p_col_name,p_col_data,p_col_qualidade):
    x=p_col_data.to_numpy()
    y=p_col_qualidade
    model = LinearRegression()
    model.fit(x.reshape((-1,1)), y)
    r_sq = model.score(x.reshape((-1,1)), y)
    print('coefficient of determination:', r_sq)
    m=model.coef_
    b=model.intercept_
    print('m (model.coef_)=',model.coef_)
    print('b (model.intercept_)=',model.intercept_)
    
    plt.plot(x,y,'.')
    plt.plot(x,x*m+b,label=(str(m)+'*x+'+str(b)))
    plt.legend(loc='upper left')
    plt.show()
```

```python
#The minimum or lowest value of the dataset
#The first quartile Q1, which represents a quarter of the way through the list of all data
#The median of the data set, which represents the midpoint of the whole list of data
#The third quartile Q3, which represents three-quarters of the way through the list of all data
#The maximum or highest value of the data set.
#
# p_verbose 0 to N
def analysis_iqr(p_col_name,p_col_data,p_verbose):
    if p_verbose>1:
        # Using Pandas
        q=p_col_data.quantile([0.25,0.5,0.75])
        print('Using Pandas quantile:\n',q)
        # Using NumPy
        q1=np.quantile(p_col_data,0.25)  # Q1
        q2=np.quantile(p_col_data,0.50)  # median
        q3=np.quantile(p_col_data,0.75)  # Q3
        print('Using np.quantile: '+str(q1)+'  '+str(q2)+'  '+str(q3))
        # Using Pandas describe"
        print('Describe:')
        print(p_col_data.describe())
    
        # Show the data
        print('The 5 Numbers:')
        print('Min:  %.5f' % np.min(p_col_data))
        print('Q1:   %.5f' % q1)
        print('Mean: %.5f' % np.mean(p_col_data))
        print('Q3:   %.5f' % q2)
        print('Max:  %.5f' % np.max(p_col_data))
        print('--------------------------')
    q1=np.quantile(p_col_data,0.25)  # Q1
    q3=np.quantile(p_col_data,0.75)  # Q3
    iqr=q3-q1
    fence1=q1-iqr*3.0
    fence2=q1-iqr*1.5
    fence3=q3+iqr*1.5
    fence4=q3+iqr*3.0
    
    total_outliers_lower=0
    total_outliers_upper=0
    for v in p_col_data:
        if v<fence2:
            if p_verbose>0:
                print('Lower Outlier: %.3f'%v)
            total_outliers_lower=total_outliers_lower+1
        if v>fence3:
            if p_verbose>0:
                print('Upper Outlier: %.3f'%v)
            total_outliers_upper=total_outliers_upper+1
    return '%s\t- Fences:  %.3f |  %.3f |  %.3f | %.3f \t Total of Outliers: %d and %d'%(p_col_name,fence1,fence2,fence3,fence4,total_outliers_lower,total_outliers_upper)
```
