```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import os
import sys
sys.path.insert(0, os.path.abspath('../'))
import rhplib.datasets_pandas
```

```python
#load data from a CSV
dataset = pd.read_csv('../../datasets/animals.csv', header=0, sep=';')
dataset.head()
```

```python
# Check the existance of null values
dataset.isnull().sum()
```

```python
# Histogram of features
rhplib.datasets_pandas.plot_hist_dataset(dataset,2,2,False)
```

```python
# Violin
sns.set_theme(style="whitegrid")
```

```python
"""
This type of diagram enable us to conclude, for example, 
that when the animal dont has legs, he always lives in the water. 

On the other hand, if has, the majority of the animals dont live in the water.
"""
```

```python
for c in dataset.columns:
    if c not in ['name','lives in the water']:
        fig = plt.figure()
        ax = sns.violinplot(x=c,y='lives in the water',data=dataset)
```

```python
"""
Here the data is grouped (by hue), in the example, if can swim or not.
In the case of animals that cannot swim, but have legs, they live always out of the water.
In the opposite case, can swim but do not have legs, they lives allways in the water.

Note that the case of both conditions not being verified there is no animals.
In the fourth case of the first diagram, if both conditions are verified, the majority of the animals 
that dont have legs lives outside of the water. 
"""
for c in dataset.columns:
    if c not in ['name','lives in the water','Can swim']:
        fig = plt.figure()
        ax = sns.violinplot(x=c,y='lives in the water',data=dataset,hue='Can swim')
```
